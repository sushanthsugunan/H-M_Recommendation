# Exploratory Data Analysis

```python
# Importing all the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Read the files
articles = pd.read_csv('articles.csv')
customers = pd.read_csv('customers.csv')
transactions = pd.read_csv('transactions_train.csv')
```

### Customer Data

```python
# get the statistics of customer data
customers.describe()
```

```python
# Category-wise count of customer club status
x = pd.DataFrame(customers['club_member_status'].value_counts())
sns.barplot(x=x.club_member_status,y=x.index)
```

```python
# percentage of fashhion news subscribers in data
sns.countplot(customers['fashion_news_frequency'])
```

### Article data

```python
A = articles['product_group_name'].value_counts()/articles['product_group_name'].shape[0]*100
A = pd.DataFrame(A)
sns.barplot(x=A.product_group_name,y=A.index)
```

```python
# Product group count in data
sns.countplot(y=articles['product_group_name'])
```

```python
#Ploting 'index_group_name Vs Count
f, ax = plt.subplots(figsize=(10, 6))
ax = sns.histplot(data=articles, y='index_group_name', color='orange')
ax.set_xlabel('count by index name')
ax.set_ylabel('index group name')
plt.show()
```
